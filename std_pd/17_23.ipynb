### 17. Compute the mean squared error on a truth and predicted series

```python
truth = pd.Series(range(10))
pred = pd. Series(range(10)) + np.random.random(10)

print(truth-pred)
print(np.mean((truth-pred)**2))
```

### 18. Convert the first character of each element in a series to uppercase

```python
ser = pd.Series(['how','to','kick','ass?'])

# sol1
ser.map(lambda x: x.title())

# sol2
pd.Series([i.title() for i in ser])
```

### 20. Compute difference of differences between consequtive numbers of a series

```python
ser = pd.Series([1,3,6,10,15,21,27,35])

print(ser.diff().tolist())
print(ser.diff().diff().tolist())
```

### 21. Convert a series of date-strings to timeseries

```python
ser = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])

pd.to_datetime(ser)
```

### 22. Get the day of month, week number, day of year and day of week from a series of data strings?

```python
ser = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])

ser_ts = pd.to_datetime(ser)

# day of month
print("Date:", ser_ts.dt.day.tolist())

# week number
print("Week num:", ser_ts.dt.weekofyear.tolist())

# day of year
print("Day of year:", ser_ts.dt.dayofyear.tolist())

# day of week
print("Day of week:", list(ser_ts.dt.day_name()))
```
